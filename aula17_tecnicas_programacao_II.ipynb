# 🧠 O que é Controle de Versão?

Controle de versão é um sistema que permite:

- registrar alterações no código
- voltar para versões anteriores
- trabalhar em equipe
- evitar perda de código



# 📚 O que é Git?

Git é um sistema de controle de versão distribuído.

Ele permite:

- registrar alterações
- comparar versões
- trabalhar com branches
- colaborar com outros desenvolvedores



# 🌐 O que é GitHub?

GitHub é uma plataforma online para hospedar repositórios Git.

Com ele podemos:

- publicar projetos
- colaborar com outros desenvolvedores
- compartilhar código
- manter portfólio profissional



# ⚙ Instalação do Git

Baixe o Git em:

https://git-scm.com



# 🚀 Inicializando um Repositório

Dentro da pasta do projeto execute:


````python
git init
````


Esse comando cria um repositório Git no projeto.



# 📄 Verificando Status


````python
git status
````


Esse comando mostra:

- arquivos modificados
- arquivos novos
- arquivos prontos para commit



# ➕ Adicionando Arquivos


````python
git add .
````


Esse comando adiciona todos os arquivos ao stage.



# 💾 Criando um Commit


````python
git commit -m "Primeira versão do sistema de produtos"
````


Um commit registra uma alteração no histórico do projeto.



# 🌿 Trabalhando com Branches


````python
git branch nova-feature
git checkout nova-feature
````


Branches permitem desenvolver funcionalidades sem afetar a versão principal.



# 🔗 Conectando com GitHub


````python
git remote add origin https://github.com/usuario/repositorio.git
````


# 📤 Enviando Projeto para GitHub


````python
git push -u origin main
````


Agora seu projeto estará disponível no GitHub.
