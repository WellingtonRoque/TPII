# 🎨 O que é Bootstrap?

Bootstrap é um **framework CSS** que facilita a criação de interfaces modernas.

Ele oferece:

- layout responsivo
- componentes prontos
- botões
- menus
- formulários
- tabelas



Site oficial:

https://getbootstrap.com

Podemos utilizar Bootstrap adicionando este link dentro do HTML:


````python
<link
href="https://cdn.jsdelivr.net/npm/bootstrap@5.3.2/dist/css/bootstrap.min.css"
rel="stylesheet">
````


# 📄 Criando um Layout Base

Em aplicações web normalmente criamos um **layout base** que será reutilizado.

Crie o arquivo:

```
templates/base.html
```


````python
<!DOCTYPE html>
<html>
<head>

<title>Sistema de Produtos</title>

<link href="https://cdn.jsdelivr.net/npm/bootstrap@5.3.2/dist/css/bootstrap.min.css" rel="stylesheet">

</head>

<body>

<nav class="navbar navbar-dark bg-dark">
    <div class="container">
        <a class="navbar-brand" href="/">Sistema de Produtos</a>
    </div>
</nav>

<div class="container mt-4">

{% block conteudo %}

{% endblock %}

</div>

</body>
</html>
````


Observe que criamos um bloco chamado:

```
{% block conteudo %}
```



# 📄 Criando a Página de Produtos

Agora vamos criar:

```
templates/produtos.html
```


````python
{% extends "base.html" %}

{% block conteudo %}

<h1>Lista de Produtos</h1>

<ul class="list-group">

{% for produto in produtos %}
<li class="list-group-item">{{ produto }}</li>
{% endfor %}

</ul>

{% endblock %}
````


Observe que utilizamos:

```
{% extends "base.html" %}
```



Isso significa que **a página utiliza o layout base**.



# 🧠 Atualizando o Controller


````python
from flask import render_template

def produto_routes(app):

    produtos = [
        "Notebook",
        "Mouse",
        "Teclado"
    ]

    @app.route("/produtos")
    def listar_produtos():
        return render_template("produtos.html", produtos=produtos)
````


Agora o sistema possui:

- layout profissional
