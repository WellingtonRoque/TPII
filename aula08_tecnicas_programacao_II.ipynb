# 🔎 O que é uma Busca?

Busca é uma **consulta filtrada** no banco de dados.

Exemplo:

Usuário digita:

```
notebook
```

O sistema retorna apenas produtos que possuem **notebook no nome**.



# 📄 Criando Formulário de Busca

Vamos adicionar um formulário no topo da página de produtos.

Arquivo:

```
templates/produtos.html
```


````python
<form method="GET" action="/buscar">

<input type="text" name="nome" placeholder="Buscar produto">

<button type="submit">Buscar</button>

</form>
````


Esse formulário enviará o termo de busca para a rota:

```
/buscar
```



# 🧠 Criando a Rota de Busca


````python
from flask import request

@app.route("/buscar")
def buscar():

    termo = request.args.get("nome")

    produtos = Produto.query.filter(
        Produto.nome.contains(termo)
    ).all()

    return render_template("produtos.html", produtos=produtos)
````


# 🔍 Explicando o Código

### request.args.get()

Obtém o valor enviado no formulário.

### contains()

Filtra produtos que contenham o termo informado.
