# 📚 O que é Paginação?

Paginação é a técnica de dividir uma lista de resultados em **várias páginas menores**.

Exemplo:

Página 1 → produtos 1 a 10  
Página 2 → produtos 11 a 20  
Página 3 → produtos 21 a 30



# ⚠ Problema Sem Paginação

Imagine que o banco possua **500 produtos**.

Se fizermos:

```
Produto.query.all()
```

Todos os produtos serão carregados de uma vez.


````python
produtos = Produto.query.all()

for p in produtos:
    print(p.nome)
````


Isso pode causar:

- carregamento lento
- excesso de informação na tela



# 🚀 Paginação com SQLAlchemy

SQLAlchemy possui suporte nativo para paginação.


````python
pagina = request.args.get("page", 1, type=int)

produtos = Produto.query.paginate(
    page=pagina,
    per_page=5
)
````


### Explicação

- **page** → número da página atual
- **per_page** → quantidade de registros por página



# 🧠 Atualizando o Controller


````python
@app.route("/produtos")
def listar_produtos():

    pagina = request.args.get("page", 1, type=int)

    produtos = Produto.query.paginate(
        page=pagina,
        per_page=5
    )

    return render_template("produtos.html", produtos=produtos)
````


# 📄 Atualizando o Template


````python
{% for p in produtos.items %}

<tr>
<td>{{ p.nome }}</td>
<td>{{ p.preco }}</td>
</tr>

{% endfor %}
````


Observe que utilizamos:

```
produtos.items
```

Isso retorna apenas os produtos da página atual.



# 🔢 Criando Navegação de Páginas


````python
<div>

{% if produtos.has_prev %}
<a href="/produtos?page={{ produtos.prev_num }}">Anterior</a>
{% endif %}

{% if produtos.has_next %}
<a href="/produtos?page={{ produtos.next_num }}">Próximo</a>
{% endif %}

</div>
````


Agora o sistema permitirá navegar entre páginas.
