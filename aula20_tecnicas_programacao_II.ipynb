## 🛒 Sistema de Catálogo de Produtos

Chegamos à última aula da disciplina.  
Durante o semestre os alunos desenvolveram um **sistema web completo** utilizando:

- Python
- Flask
- SQLAlchemy
- Bootstrap
- Padrões de Projeto
- Testes
- Git e GitHub

Nesta aula cada grupo irá **apresentar o projeto desenvolvido**.



# 🎯 Objetivos da Aula

Ao final desta aula os alunos deverão:

- apresentar o sistema desenvolvido
- explicar as tecnologias utilizadas
- demonstrar as funcionalidades do projeto
- refletir sobre os desafios encontrados durante o desenvolvimento



# 📋 Estrutura da Apresentação

Cada grupo terá aproximadamente **10 a 15 minutos** para apresentar.

A apresentação deve conter:

1. Introdução do projeto
2. Problema que o sistema resolve
3. Demonstração do sistema
4. Tecnologias utilizadas
5. Desafios enfrentados
6. Melhorias futuras



# 🧱 Demonstração do Sistema

Durante a apresentação o grupo deve mostrar:

- execução da aplicação
- cadastro de produtos
- edição de produtos
- exclusão de produtos
- busca de produtos
- paginação



# 💻 Execução do Sistema

Passos para executar o sistema:

```
pip install -r requirements.txt
python app.py
```

Acesse no navegador:

```
http://localhost:5000
```
