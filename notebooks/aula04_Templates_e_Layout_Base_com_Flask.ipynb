# 🧱 Layout Base (PARTE PRINCIPAL DA AULA)

Em sistemas reais NÃO repetimos HTML em todas as páginas.

Problema:

- repetir header em todas páginas ❌  
- repetir menu ❌  
- repetir estrutura ❌  

Solução:

👉 Criar um **layout base reutilizável**


# 🧠 Conceito de Layout Base

O layout base funciona como um **molde da aplicação**.

Ele define:

- cabeçalho  
- menu  
- estrutura geral  

E deixa espaços para conteúdo dinâmico.


# 📄 Criando o base.html

```html
<!DOCTYPE html>
<html>

<head>
    <title>Sistema de Produtos</title>
</head>

<body>

<header>
    <h1>Sistema de Produtos</h1>
</header>

<nav>
    <a href="/">Home</a>
    <a href="/produtos">Produtos</a>
</nav>

<hr>

{% block conteudo %}
{% endblock %}

</body>
</html>
```


# 🔥 Entendendo o bloco

```html
{% block conteudo %}
{% endblock %}
```

👉 Isso é um espaço dinâmico

👉 Outras páginas vão inserir conteúdo aqui


# 📄 Criando página que herda o layout

```html
{% extends "base.html" %}

{% block conteudo %}

<h2>Lista de Produtos</h2>

<p>Aqui vão aparecer os produtos</p>

{% endblock %}
```


# 🧠 O que está acontecendo

- extends → herda o layout  
- block → substitui conteúdo  

👉 Flask monta a página automaticamente


# 🧪 Resultado Final (mental)
