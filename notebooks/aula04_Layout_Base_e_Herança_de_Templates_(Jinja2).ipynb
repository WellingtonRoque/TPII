# 🧠 Problema Real

Sem layout base:

- HTML duplicado ❌
- difícil manutenção ❌
- inconsistência visual ❌

👉 Em sistemas grandes isso vira caos


# 📄 Base simples (inicial)

```html
<!DOCTYPE html>
<html>
<body>

<header>
    <h1>Sistema</h1>
</header>

{% block conteudo %}{% endblock %}

</body>
</html>
```


# 🚀 Evolução — Múltiplos Blocos (IMPORTANTE)

Agora vamos evoluir para algo PROFISSIONAL:

```html
<!DOCTYPE html>
<html>

<head>
    <title>Sistema</title>
</head>

<body>

{% block header %}
<header>
    <h1>Sistema de Produtos</h1>
</header>
{% endblock %}

{% block menu %}
<nav>
    <a href="/">Home</a>
    <a href="/produtos">Produtos</a>
    <a href="/clientes">Clientes</a>
</nav>
{% endblock %}

<hr>

{% block conteudo %}
{% endblock %}

<hr>

{% block footer %}
<footer>
    <p>2026 - Sistema</p>
</footer>
{% endblock %}

</body>
</html>
```


# 🧠 O que fizemos aqui

Agora temos:

- block header → topo
- block menu → navegação
- block conteudo → conteúdo da página
- block footer → rodapé

👉 Isso é padrão de sistema real


# 📄 Sobrescrevendo blocos

Podemos alterar partes específicas:

```html
{% extends "base.html" %}

{% block conteudo %}
<h2>Produtos</h2>
{% endblock %}
```

👉 só muda o conteúdo


# 📄 Alterando apenas o menu

```html
{% extends "base.html" %}

{% block menu %}
<nav>
    <a href="/">Home</a>
    <a href="/produtos">Produtos</a>
    <a href="/admin">Admin</a>
</nav>
{% endblock %}

{% block conteudo %}
<h2>Painel Admin</h2>
{% endblock %}
```

👉 substituímos apenas um bloco específico


# 🧱 Estrutura do Projeto

```
templates/
    base.html
    index.html
    produtos.html
    clientes.html
    admin.html
```


````python
from flask import Flask, render_template

app = Flask(__name__)

@app.route("/")
def home():
    return render_template("index.html")

@app.route("/produtos")
def produtos():
    return render_template("produtos.html")

@app.route("/clientes")
def clientes():
    return render_template("clientes.html")

@app.route("/admin")
def admin():
    return render_template("admin.html")

if __name__ == "__main__":
    app.run(debug=True)
````
