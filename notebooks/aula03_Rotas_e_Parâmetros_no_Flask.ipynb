Agora vamos tornar o sistema dinâmico.


```python
from flask import Flask

app = Flask(__name__)

@app.route("/")
def home():
    return "Sistema de Produtos"

if __name__ == "__main__":
    app.run(debug=True)
```

# 🧠 Rotas Dinâmicas

Permitem receber valores pela URL.

Exemplo:
/produto/notebook


```python
@app.route("/produto/<nome>")
def produto(nome):
    return f"Produto: {nome}"
```

# 🔢 Tipos de Parâmetros

| Tipo   | Uso |
|--------|-----|
| string | textos |
| int    | números inteiros |
| float  | valores decimais |
| path   | caminhos completos |


```python
@app.route("/pedido/<int:id>")
def pedido(id):
    return f"Pedido: {id}"
```

```python
@app.route("/preco/<float:valor>")
def preco(valor):
    return f"Preço: {valor}"
```

```python
@app.route("/arquivo/<path:caminho>")
def arquivo(caminho):
    return f"Caminho: {caminho}"
```

# 🧮 Aplicando Cálculos

```python
@app.route("/soma/<int:a>/<int:b>")
def soma(a, b):
    return f"Soma: {a+b}"

@app.route("/media/<float:a>/<float:b>")
def media(a, b):
    return f"Média: {(a+b)/2}"
```

# 🧠 Regras de Negócio

```python
@app.route("/desconto/<float:preco>/<int:percentual>")
def desconto(preco, percentual):
    valor = preco - (preco * percentual / 100)
    return f"Preço com desconto: {valor}"
```
