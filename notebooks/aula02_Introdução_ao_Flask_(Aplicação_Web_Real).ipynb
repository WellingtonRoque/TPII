# 🧠 O que é uma aplicação web?

Uma aplicação web é um sistema que roda em um servidor e responde requisições feitas por um navegador.

Fluxo:

Usuário → Navegador → Servidor Flask → Resposta

👉 Agora nosso sistema de produtos vai seguir esse modelo.


# ⚙️ Instalação do Flask

No terminal:

```

```

Verifique:

```

```


# 🧱 Estrutura inicial do projeto

```
projeto/
    app.py
```


````python
from flask import Flask

app = Flask(__name__)

@app.route("/")
def home():
    return "Sistema de Produtos"

if __name__ == "__main__":
    app.run(debug=True)
````

# 🧪 Execução

```
python app.py
```

Acesse:

http://127.0.0.1:5000


# 🔍 Entendendo o código

- Flask cria o servidor  
- @app.route define a URL  
- def home() responde a requisição  


# 🧱 Evoluindo o sistema (estrutura real)

Nosso sistema precisa de páginas:

- produtos  
- clientes  
- pedidos  
- categorias  


````python
@app.route("/produtos")
def produtos():
    return "Lista de produtos"

@app.route("/clientes")
def clientes():
    return "Lista de clientes"

@app.route("/pedidos")
def pedidos():
    return "Lista de pedidos"

@app.route("/categorias")
def categorias():
    return "Lista de categorias"
````
