# 🧠 Controller (IMPORTANTE)

Agora o Controller começa a trabalhar com dados reais do sistema.


````python
from flask import Flask, render_template

app = Flask(__name__)

produtos = [
    {"nome": "Notebook", "preco": 3000},
    {"nome": "Mouse", "preco": 100}
]

@app.route("/produtos")
def produtos_view():
    return render_template("produtos.html", produtos=produtos)

if __name__ == "__main__":
    app.run(debug=True)
````

# 📄 Tela de Produtos (REAL)

```html
{% extends "base.html" %}

{% block conteudo %}

<h2>Lista de Produtos</h2>

<table class="table table-striped">
    <thead>
        <tr>
            <th>Nome</th>
            <th>Preço</th>
        </tr>
    </thead>

    <tbody>
        {% for p in produtos %}
        <tr>
            <td>{{ p.nome }}</td>
            <td>{{ p.preco }}</td>
        </tr>
        {% endfor %}
    </tbody>
</table>

<a href="/cadastro" class="btn btn-primary">Cadastrar Produto</a>

{% endblock %}
```


# 📄 Tela de Cadastro (REAL)

```html
{% extends "base.html" %}

{% block conteudo %}

<h2>Cadastrar Produto</h2>

<form method="POST">

    <div class="mb-3">
        <label>Nome</label>
        <input type="text" name="nome" class="form-control">
    </div>

    <div class="mb-3">
        <label>Preço</label>
        <input type="number" name="preco" class="form-control">
    </div>

    <button class="btn btn-success">Salvar</button>

</form>

{% endblock %}
```


````python
from flask import request, redirect

@app.route("/cadastro", methods=["GET", "POST"])
def cadastro():

    if request.method == "POST":
        nome = request.form["nome"]
        preco = float(request.form["preco"])

        produtos.append({"nome": nome, "preco": preco})

        return redirect("/produtos")

    return render_template("cadastro_produto.html")
````

# 🔥 Fluxo completo do sistema

1. Usuário acessa /produtos  
2. Controller envia lista  
3. Bootstrap renderiza tabela  
4. Usuário clica em cadastrar  
5. Formulário envia dados  
6. Controller processa  
7. Produto é adicionado  

👉 Isso é sistema real funcionando
