# 🌐 O que é uma Requisição HTTP?

Quando acessamos um site, o navegador envia uma requisição para o servidor.

Exemplos:

```
GET /produtos
POST /novo
```



# 🚀 Flask Test Client

Flask possui uma ferramenta chamada:

```
test_client()
```

Ela permite simular requisições HTTP.


````python
from app import app

client = app.test_client()

response = client.get("/produtos")

print(response.status_code)
````


Se a rota existir, o código retornado normalmente será:

```
200
```



# 🧪 Criando um Teste de Rota


````python
import unittest
from app import app

class TestRotas(unittest.TestCase):

    def setUp(self):
        self.client = app.test_client()

    def test_pagina_produtos(self):

        response = self.client.get("/produtos")

        self.assertEqual(response.status_code,200)
````


Esse teste verifica se a rota **/produtos** está funcionando.



# 🧪 Testando Cadastro de Produto


````python
def test_criar_produto(self):

    response = self.client.post("/novo", data={
        "nome":"Produto Teste",
        "preco":"100"
    })

    self.assertEqual(response.status_code,302)
````


O código **302** indica redirecionamento após cadastro.
