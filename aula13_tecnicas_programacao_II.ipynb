# ⚠ Problema Sem Repository

Sem o padrão, o controller acessa o banco diretamente:


````python
@app.route("/produtos")
def listar_produtos():

    produtos = Produto.query.all()

    return render_template("produtos.html", produtos=produtos)
````


Se vários controllers fizerem isso:

- o código fica repetitivo
- fica difícil alterar a forma de acesso ao banco



# 🚀 Solução — Repository Pattern

Criamos uma classe responsável por **todas as operações com o banco**.



# 📂 Nova Estrutura do Projeto

```
projeto/

models/
    produto.py

repositories/
    produto_repository.py

controllers/
    produto_controller.py
```



# 🧱 Criando o Repository


````python
# repositories/produto_repository.py

class ProdutoRepository:

    def listar(self):
        return Produto.query.all()

    def buscar_por_id(self, id):
        return Produto.query.get(id)

    def salvar(self, produto):
        db.session.add(produto)
        db.session.commit()

    def excluir(self, produto):
        db.session.delete(produto)
        db.session.commit()
````


Agora todas as operações com o banco ficam centralizadas.



# 🧠 Utilizando o Repository no Controller


````python
repo = ProdutoRepository()

@app.route("/produtos")
def listar_produtos():

    produtos = repo.listar()

    return render_template("produtos.html", produtos=produtos)
````


Observe que agora o controller **não acessa diretamente o banco**.



# ✏ Atualizando Cadastro de Produto


````python
repo = ProdutoRepository()

@app.route("/novo", methods=["GET","POST"])
def novo_produto():

    if request.method == "POST":

        nome = request.form["nome"]
        preco = request.form["preco"]

        produto = Produto(nome=nome, preco=preco)

        repo.salvar(produto)

        return redirect("/produtos")

    return render_template("novo_produto.html")
````
