# 🌐 O que são Rotas?

Rotas conectam **URLs acessadas pelo navegador** com **funções Python**.

Exemplo:

```
URL → http://localhost:5000/produtos
Função → produtos()
```

Ou seja:

Quando o usuário acessa `/produtos`, o Flask executa a função correspondente.



# 🧱 Criando múltiplas rotas


````python
from flask import Flask

app = Flask(__name__)

@app.route("/")
def home():
    return "Página inicial"

@app.route("/sobre")
def sobre():
    return "Página sobre o sistema"

@app.route("/produtos")
def produtos():
    return "Lista de produtos"

@app.route("/contato")
def contato():
    return "Página de contato"

if __name__ == "__main__":
    app.run(debug=True)
````

# 🧠 Rotas com parâmetros

Uma das grandes vantagens do Flask é permitir **rotas dinâmicas**.

Exemplo:

```
/produto/notebook
/produto/mouse
/produto/teclado
```


````python
@app.route("/produto/<nome>")
def produto(nome):
    return f"Produto selecionado: {nome}"
````


Se o usuário acessar:

```
http://localhost:5000/produto/notebook
```

A saída será:

```
Produto selecionado: notebook
```



# 🔢 Parâmetros numéricos

Também podemos restringir o tipo do parâmetro.


````python
@app.route("/produto/id/<int:id>")
def produto_id(id):
    return f"ID do produto: {id}"
````


Tipos de parâmetros suportados:

| Tipo | Descrição |
|-----|-----------|
string | texto |
int | número inteiro |
float | número decimal |
path | caminho completo |



# 🧪 Exemplo completo com parâmetros


````python
from flask import Flask

app = Flask(__name__)

@app.route("/")
def home():
    return "Sistema de Produtos"

@app.route("/usuario/<nome>")
def usuario(nome):
    return f"Bem-vindo {nome}"

@app.route("/produto/<nome>")
def produto(nome):
    return f"Produto: {nome}"

@app.route("/pedido/<int:id>")
def pedido(id):
    return f"Número do pedido: {id}"

if __name__ == "__main__":
    app.run(debug=True)
````
