## 🛒 Sistema de Catálogo de Produtos  

Nesta aula vamos evoluir significativamente o sistema, saindo de rotas simples para um sistema dinâmico capaz de processar dados, validar entradas e aplicar regras de negócio.

👉 Essa aula é fundamental para preparar o sistema para banco de dados e CRUD.


# 🎯 Objetivos

- Dominar rotas no Flask
- Trabalhar com parâmetros dinâmicos
- Utilizar tipos: string, int, float e path
- Aplicar cálculos reais no sistema
- Entender erros comuns
- Evoluir o sistema com lógica de negócio


```python
from flask import Flask

app = Flask(__name__)

@app.route("/")
def home():
    return "Sistema de Produtos"

if __name__ == "__main__":
    app.run(debug=True)
```

# 🧱 Estrutura do Sistema

Vamos criar as páginas principais do sistema:

- produtos
- clientes
- pedidos
- categorias


```python
@app.route("/produtos")
def produtos():
    return "Lista de produtos"

@app.route("/clientes")
def clientes():
    return "Lista de clientes"

@app.route("/pedidos")
def pedidos():
    return "Lista de pedidos"

@app.route("/categorias")
def categorias():
    return "Lista de categorias"
```

# 🧠 Parâmetros STRING

```python
@app.route("/produto/<nome>")
def produto(nome):
    return f"Produto: {nome}"
```

# 🔢 Parâmetros INT

```python
@app.route("/pedido/<int:id>")
def pedido(id):
    return f"Pedido número: {id}"
```

# 💰 Parâmetros FLOAT

```python
@app.route("/preco/<float:valor>")
def preco(valor):
    return f"Preço: {valor}"
```

# 📁 Parâmetros PATH

```python
@app.route("/arquivo/<path:caminho>")
def arquivo(caminho):
    return f"Caminho: {caminho}"
```

# 🧮 Aplicando Cálculos no Sistema

```python
@app.route("/soma/<int:a>/<int:b>")
def soma(a,b):
    return f"Soma: {a+b}"

@app.route("/subtracao/<int:a>/<int:b>")
def sub(a,b):
    return f"Subtração: {a-b}"

@app.route("/multiplicacao/<int:a>/<int:b>")
def mult(a,b):
    return f"Multiplicação: {a*b}"

@app.route("/divisao/<float:a>/<float:b>")
def div(a,b):
    return f"Divisão: {a/b}"
```

# 🧠 Regras de Negócio

```python
@app.route("/desconto/<float:preco>/<int:percentual>")
def desconto(preco, percentual):
    valor = preco - (preco * percentual / 100)
    return f"Preço com desconto: {valor}"
```

```python
@app.route("/lucro/<float:custo>/<float:venda>")
def lucro(custo, venda):
    return f"Lucro: {venda - custo}"
```
