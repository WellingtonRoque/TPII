# 🧩 Problema Inicial

Imagine que precisamos desenvolver um pequeno sistema para uma loja.

Esse sistema deve permitir:

- cadastrar produtos
- listar produtos
- buscar produtos
- remover produtos


```python
produtos = []

while True:

    print("1 - Adicionar produto")
    print("2 - Listar produtos")
    print("3 - Buscar produto")
    print("4 - Remover produto")
    print("5 - Sair")

    opcao = input("Escolha: ")

    if opcao == "1":
        nome = input("Nome do produto: ")
        produtos.append(nome)

    elif opcao == "2":
        for p in produtos:
            print(p)

    elif opcao == "3":
        busca = input("Buscar produto: ")
        for p in produtos:
            if busca == p:
                print("Produto encontrado")

    elif opcao == "4":
        remover = input("Remover produto: ")
        if remover in produtos:
            produtos.remove(remover)

    elif opcao == "5":
        break
```

# 🧠 Arquitetura de Software

Arquitetura de software define **como um sistema é organizado**.

Um dos padrões mais utilizados é o **MVC**.



# MVC

| Componente | Responsabilidade |
|------------|-----------------|
| Model | dados |
| Controller | regras |
| View | interface |


```python
produtos = []

def adicionar_produto(nome):
    produtos.append(nome)

def listar_produtos():
    for p in produtos:
        print(p)

def buscar_produto(nome):
    return nome in produtos

def remover_produto(nome):
    if nome in produtos:
        produtos.remove(nome)
```

```python
def menu():

    while True:

        print("1 - Adicionar")
        print("2 - Listar")
        print("3 - Buscar")
        print("4 - Remover")
        print("5 - Sair")

        opcao = input()

        if opcao == "1":
            nome = input("Produto: ")
            adicionar_produto(nome)

        elif opcao == "2":
            listar_produtos()

        elif opcao == "3":
            nome = input("Buscar: ")
            if buscar_produto(nome):
                print("Produto encontrado")
            else:
                print("Produto não encontrado")

        elif opcao == "4":
            nome = input("Remover: ")
            remover_produto(nome)

        elif opcao == "5":
            break
```

```python
menu()
```
