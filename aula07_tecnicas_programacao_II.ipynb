```
    preco = db.Column(db.Float)
```



# 🧠 O que é CRUD?

CRUD é um acrônimo para as quatro operações básicas de manipulação de dados:

| Operação | Descrição |
|--------|-----------| Create | Criar registros | Read | Ler registros | Update | Atualizar registros | Delete | Remover registros |



# 📖 READ — Listar Produtos


````python
@app.route("/produtos")
def listar_produtos():

    produtos = Produto.query.all()

    return render_template("produtos.html", produtos=produtos)
````


# ➕ CREATE — Cadastrar Produto


````python
@app.route("/novo", methods=["GET","POST"])
def novo_produto():

    if request.method == "POST":

        nome = request.form["nome"]
        preco = request.form["preco"]

        produto = Produto(nome=nome, preco=preco)

        db.session.add(produto)
        db.session.commit()

        return redirect("/produtos")

    return render_template("novo_produto.html")
````


# 📄 Formulário HTML


````python

<form method="POST">

<input name="nome" placeholder="Nome do produto">

<input name="preco" placeholder="Preço">

<button type="submit">Salvar</button>

</form>
````


# ✏ UPDATE — Editar Produto


````python
@app.route("/editar/<id>", methods=["GET","POST"])
def editar_produto(id):

    produto = Produto.query.get(id)

    if request.method == "POST":

        produto.nome = request.form["nome"]
        produto.preco = request.form["preco"]

        db.session.commit()

        return redirect("/produtos")

    return render_template("editar_produto.html", produto=produto)
````


# ❌ DELETE — Excluir Produto


````python
@app.route("/excluir/<id>")
def excluir_produto(id):

    produto = Produto.query.get(id)

    db.session.delete(produto)

    db.session.commit()

    return redirect("/produtos")
````
